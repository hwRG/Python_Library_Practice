#  * 별첨 : 위 데이터 분석에 사용된 전체 코드  *
사용자가 직접 입력해 원하는 작업을 수행할 수 있습니다.<br>
처음에 Yes와 No 두 방향으로 나뉘는데, Yes를 입력할 시 전국 데이터를 가져와 전국 단위로 GDP 값을 비교할 수 있습니다.<br>
No를 입력할 시 특정 지역의 도시간의 인구와 GDP값을 비교할 수 있습니다.

```python
import csv
import time

import matplotlib.pyplot as plt
import numpy as np
 
from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)    

# 딕셔너리에 각 지역 define
regionDic = {0 : '서울특별시', 1 : '경기도', 2 : '인천광역시', 3 : '부산광역시', 4 : '대전광역시', 5 : '대구광역시', 6 : '광주광역시', 7 : '울산광역시', 8 : '충청북도', 9 : '충청남도', 10 : '전라북도', 11 : '전라남도', 12 : '경상북도', 13 : '경상남도'}  

# 초기화면 구성
while True :    
    print("\n-----------------------------------------------")
    print("If you choose no, you will be compared by city.")
    print("If you want to stop, press just Enter")
    userWishAllRegion = input("Will you call all regions?(Yes or No) : ")
    print("-----------------------------------------------")

    # 지역별로 검색 후 지정한 산업의 가장 높은 수치의 도시 찾아내기
    if userWishAllRegion == 'Yes' or userWishAllRegion == 'yes' or userWishAllRegion == '응' :
        allPopulation = []
        tempPop = []

        print("")
        # 모든 지역의 인구 수를 리스트에 저장 (3중 리스트)
        for i in range(len(regionDic)) :
            with open(str(regionDic[i]) + "_주민등록인구및세대현황_월간.csv","r") as nationWide_Population_f :
                for lineContent in nationWide_Population_f :
                    tempPop.append(lineContent.strip('\n').split(','))
                print("Succeeded in calling", str(regionDic[i]) + "_주민등록인구및세대현황_월간.csv!")
                allPopulation.append(tempPop)
                tempPop = []
            time.sleep(0.03)
            
        # 도시 문자열을 도시 이름으로 가공하는 과정 (ex.'경기도 수원시' -> '수원시')
        for k in range(len(regionDic)) :
            for i in range(len(allPopulation[k])) : 
                allPopulation[k][i][0] = allPopulation[k][i][0].replace(' ','')
                allPopulation[k][i][0] = allPopulation[k][i][0].replace(regionDic[k],'')
                allPopulation[k][i][0] = allPopulation[k][i][0].replace('(','')
                allPopulation[k][i][0] = allPopulation[k][i][0].replace(')','')
                allPopulation[k][i][0] = allPopulation[k][i][0].replace('"','')

                # 도시 문자열의 불필요한 숫자들 제거
                for j in range(10) :
                    j = str(j)
                    allPopulation[k][i][0] = allPopulation[k][i][0].replace(j,'')
                
        allGDP = []
        tempGDP = []
        print("")
        
        # 모든 지역의 GDP를 리스트에 저장 (3중 리스트)
        for i in range(len(regionDic)) :
            with open(str(regionDic[i]) + "_경제활동별_지역내총생산.csv","r") as nationWide_GDP_f :
                for lineContent in nationWide_GDP_f :
                    tempGDP.append(lineContent.strip('\n').split(','))
                print("Succeeded in calling", str(regionDic[i]) + "_경제활동별_지역내총생산.csv!")
                allGDP.append(tempGDP)
                tempGDP = []
            time.sleep(0.03)

        # 도시 문자열을 도시 이름으로 가공 (ex.'경기도 수원시' -> '수원시')
        for k in range(len(regionDic)) :
            for i in range(len(allGDP[k])) : 
                allGDP[k][i][1] = allGDP[k][i][1].replace(',',' ')
                allGDP[k][i][1] = allGDP[k][i][1].replace('"','')
                allGDP[k][i][0] = allGDP[k][i][0].replace('"','')
                allGDP[k][i][1] = allGDP[k][i][1].replace('지역내총생산(시장가격)','총GDP')
                allGDP[k][i][2] = allGDP[k][i][2].replace(',',' ')
                allGDP[k][i][2] = allGDP[k][i][2].replace('"','')
                allGDP[k][i][1] = allGDP[k][i][1].replace('"','')
                allGDP[k][i][2] = allGDP[k][i][2].replace('지역내총생산(시장가격)','총GDP')

        # GDP 항목의 컴마 오류 해결    
        for k in range(len(regionDic)) :
            for i in range(len(allGDP[k])) :
                if allGDP[k][i][2] == ' 가스' or allGDP[k][i][2] == '가스' or allGDP[k][i][2] == ' 영상' or allGDP[k][i][2] == '영상' or allGDP[k][i][2] == ' 국방 및  사회보장행정' or allGDP[k][i][2] == '국방및사회보장행정' or allGDP[k][i][2] == ' 국방 및 사회보장행정' or allGDP[k][i][2] == '국방 및 사회보장행정' or allGDP[k][i][2] == '국방 및  사회보장행정' or allGDP[k][i][2] == '국방및사회보장행정': 
                    for j in range(len(allGDP[k][i])-3) :
                        allGDP[k][i][1] += ',' + allGDP[k][i][2]
                        allGDP[k][i].remove(allGDP[k][i][2])
                    allGDP[k][i][1] = allGDP[k][i][1].replace('"','')
                    allGDP[k][i][1] = allGDP[k][i][1].replace(' ','')
                    allGDP[k][i][1] = allGDP[k][i][1].replace(',','')

    
        # 지역 산업의 GDP가 0일 때 '-' 라고 표기된 것을 '0'으로 수정            
        for i in range(len(regionDic)) :
            for j in range(len(allGDP[i])) :
                if i == 5 or i == 9 :
                    if allGDP[i][j][3] == '-' :
                        allGDP[i][j][3] = '0'
                else :
                    if allGDP[i][j][2] == '-' :
                        allGDP[i][j][2] = '0'

        # 지역마다 다른 이름 구성으로 인해 띄어쓰기를 없애고 최대한 이름 통일
        for k in range(len(regionDic)) :
            for i in range(len(allGDP[k])) :    
                for j in range(len(allGDP[k][i])) :
                    allGDP[k][i][j] = allGDP[k][i][j].replace(' ','') 
            
        # 비교하고 싶은 산업 선택
        while True :
            print('\n산업 종류')
            for i in range(13) :
                print(allGDP[1][i+3][1], end=' / ')
            print(allGDP[1][i+4][1], "\n※ 도매및소매업, 출판영상방송통신및정보서비스업, 부동산및임대업은 확인 불가 ※")

            
            allHopeGDPvalue = []
            allHopeGDPname = []
            tempallHopeGDPvalue = []
            temp2allHopeGDPvalue = []
            print("\nIf you want to back, press 0")
            userHopeIndustry = input("Select the industry you want to compare. : ")
                

            # 지역 별로 지정한 산업의 가장 높은 수치를 지니고 있는 산업 선택 후 리스트 저장
            if userHopeIndustry == '0' : break
            else :
                for i in range(len(regionDic)) :
                    # 대구와 충청남도는 csv파일에서 다르게 저장되어 다르게 저장
                    if i == 5 or i == 9 :
                        for j in range(len(allGDP[i])) :
                            if allGDP[i][j][2] == userHopeIndustry :
                                tempallHopeGDPvalue.append(allGDP[i][j][3])
                            for m in range(len(tempallHopeGDPvalue)) :
                                tempallHopeGDPvalue[m] = int(tempallHopeGDPvalue[m])
                        tempallHopeGDPvalue.sort()
                        tempallHopeGDPvalue.reverse()
                        
                        temp2allHopeGDPvalue.append(tempallHopeGDPvalue[0])
                        allHopeGDPvalue.append(tempallHopeGDPvalue[0])
                        tempallHopeGDPvalue = []
                        
                        for k in range(len(allGDP[i])) :
                            for l in range(len(temp2allHopeGDPvalue)) :
                                if allGDP[i][k][3] == str(temp2allHopeGDPvalue[l]) :
                                    allHopeGDPname.append(allGDP[i][k][0])
                        temp2allHopeGDPvalue = []
                    

                    else :
                        for j in range(len(allGDP[i])) :
                            if allGDP[i][j][1] == userHopeIndustry :
                                tempallHopeGDPvalue.append(allGDP[i][j][2])
                            for m in range(len(tempallHopeGDPvalue)) :
                                tempallHopeGDPvalue[m] = int(tempallHopeGDPvalue[m])
                        tempallHopeGDPvalue.sort()
                        tempallHopeGDPvalue.reverse()

                        temp2allHopeGDPvalue.append(tempallHopeGDPvalue[0])
                        allHopeGDPvalue.append(tempallHopeGDPvalue[0])
                        tempallHopeGDPvalue = []
                        
                        for k in range(len(allGDP[i])) :
                            for l in range(len(temp2allHopeGDPvalue)) :
                                if allGDP[i][k][2] == str(temp2allHopeGDPvalue[l]) :
                                    allHopeGDPname.append(allGDP[i][k][0])
                        temp2allHopeGDPvalue = []
                

                regionNameValue = []
                tempRegionNameValue = []
                sortRegionValue = []
                sortRegionNameValue = []
                regionTopPopulation = []
                sortRegionTopPopulation = []

                # 지정한 지역 산업의 인구 수 리스트에 저장
                for i in range(len(regionDic)) :
                    for j in range(len(allPopulation[i])) :
                        if allPopulation[i][j][0] == allHopeGDPname[i] :
                            regionTopPopulation.append(allPopulation[i][j][1])
                for i in range(len(regionDic)) :
                    regionTopPopulation[i] = int(regionTopPopulation[i])

                # 나뉘어진 데이터를 각 지역의 리스트로 묶음
                for i in range(len(regionDic)) :
                    tempRegionNameValue.append(regionDic[i])
                    tempRegionNameValue.append(allHopeGDPname[i])
                    tempRegionNameValue.append(allHopeGDPvalue[i])
                    tempRegionNameValue.append(regionTopPopulation[i])
                    regionNameValue.append(tempRegionNameValue)
                    tempRegionNameValue = []
                
                
                            
                # 내림차순으로 GDP값 재정렬
                for i in range(len(regionDic)) :
                    sortRegionValue.append(regionNameValue[i][2])
                sortRegionValue.sort()
                sortRegionValue.reverse()
  
            
                # 내림차순으로 도시 이름 재정렬
                for i in range(len(regionDic)) :
                    for j in range(len(regionDic)) :
                        if sortRegionValue[i] == regionNameValue[j][2] :
                            sortRegionNameValue.append(regionNameValue[j])


                # 지역의 도시 별 1인당 GDP 값 계산
                perSortRegionValue = []
                for i in range(len(regionDic)) :
                    perSortRegionValue.append(sortRegionNameValue[i][2] / sortRegionNameValue[i][3])
                
                # 순서대로 지정한 산업 출력
                print("\n" + userHopeIndustry)
                print(regionNameValue[0], "(단위 : 백만원, 명)")
                for i in range(len(regionDic) - 1) :
                    print(regionNameValue[i+1])


                # 순서대로 지정한 산업의 1인당 GDP 출력   
                print("\n" + userHopeIndustry, "내림차순")
                print(sortRegionNameValue[0], "(단위 : 백만원, 명)")
                for i in range(len(regionDic) - 1) :
                    print(sortRegionNameValue[i+1])
                    
                
                # 산업의 GDP를 내림차순으로 정리하여 그래프 표현
                sortRegionName = []
                for i in range(len(regionDic)-1) :
                    sortRegionName.append(sortRegionNameValue[i][1])
                sortRegionName.append(sortRegionNameValue[i+1][1])
                plt.barh(sortRegionName, sortRegionValue)
                plt.ylabel('GDP')
                plt.title(userHopeIndustry)
                plt.show()
                
                # 산업의 1인당 GDP를 내림차순으로 정리하여 그래프 표현
                plt.barh(sortRegionName, perSortRegionValue)
                plt.ylabel('1인당 GDP')
                plt.title("인구대비 " + userHopeIndustry)
                plt.show()
            
            
    # 지정한 지역의 도시 두 개의 인구와 GDP 비교
    elif userWishAllRegion == 'No' or userWishAllRegion == 'no' or userWishAllRegion == '아니' :
        
        while True :
            print("\nIf you want to back, press 0")          
            userWishRegion = input("Press your hope region (ex.서울특별시 경기도 대구광역시) : ") 
            if userWishRegion == '0' :
                break
            else : 
                print("Succeeded in calling", userWishRegion + "_주민등록인구및세대현황_월간.csv!")


                allPopulation = []
                # 도시(경기도)의 인구 데이터를 csv파일에서 가져옴
                with open(userWishRegion + "_주민등록인구및세대현황_월간.csv","r") as nationWide_Population_f :
                    for lineContent in nationWide_Population_f :
                        allPopulation.append(lineContent.strip('\n').split(','))

                # 도시 문자열을 도시 이름으로 가공하는 과정 (ex.'경기도 수원시' -> '수원시')
                for i in range(len(allPopulation)) : 
                    allPopulation[i][0] = allPopulation[i][0].replace(' ','')
                    allPopulation[i][0] = allPopulation[i][0].replace(userWishRegion,'')
                    allPopulation[i][0] = allPopulation[i][0].replace('(','')
                    allPopulation[i][0] = allPopulation[i][0].replace(')','')
                    allPopulation[i][0] = allPopulation[i][0].replace('"','')

                    # 도시 문자열의 불필요한 숫자들을 제거하는 과정
                    for j in range(10) :
                        j = str(j)
                        allPopulation[i][0] = allPopulation[i][0].replace(j,'')


                # 사용자가 입력할 도시 선택
                for i in range(len(allPopulation)-1) :
                    print(allPopulation[i][0],end=' / ')
                print(allPopulation[i+1][0])


                # 지역의 총생산량 파일 불러오는 과정
                allGDP = []
                with open(userWishRegion + "_경제활동별_지역내총생산.csv","r") as nationWide_GDP_f :
                    for lineContent in nationWide_GDP_f :
                        allGDP.append(lineContent.strip('\n').split(','))

                # 도시 문자열을 도시 이름으로 가공하는 과정 (ex.'경기도 수원시' -> '수원시')
                for i in range(len(allGDP)) : 
                    allGDP[i][1] = allGDP[i][1].replace(',',' ')
                    allGDP[i][1] = allGDP[i][1].replace('"','')
                    allGDP[i][0] = allGDP[i][0].replace('"','')
                    allGDP[i][1] = allGDP[i][1].replace('지역내총생산(시장가격)','총GDP')
                    allGDP[i][2] = allGDP[i][2].replace(',',' ')
                    allGDP[i][2] = allGDP[i][2].replace('"','')
                    allGDP[i][1] = allGDP[i][1].replace('"','')
                    allGDP[i][2] = allGDP[i][2].replace('지역내총생산(시장가격)','총GDP')

                # 컴마 항목 오류 해결    
                for i in range(len(allGDP)) :
                    if allGDP[i][2] == ' 가스' or allGDP[i][2] == '가스' or allGDP[i][2] == ' 영상' or allGDP[i][2] == '영상' or allGDP[i][2] == ' 국방 및  사회보장행정' or allGDP[i][2] == '국방및사회보장행정' or allGDP[i][2] == ' 국방 및 사회보장행정'or allGDP[i][2] == '국방 및 사회보장행정' : 
                        for j in range(len(allGDP[i])-3) :
                            allGDP[i][1] += ',' + allGDP[i][2]
                            allGDP[i].remove(allGDP[i][2])
                        allGDP[i][1] = allGDP[i][1].replace('"','')
                        allGDP[i][1] = allGDP[i][1].replace(' ','')


                while True :
                    # 사용자가 희망하는 지역의 두 도시를 입력 받음
                    print("\n",userWishRegion, sep ='')
                    print("If you want to back, press 0")          
                    userFirstPosition = input("Press your hope first area : ")
                    if userFirstPosition == '0' :
                        break
                    else :
                        userSecondPosition = input("Press your hope second area : ")
                        # 선택한 도시의 인구 수 출력 과정
                        for i in range(len(allPopulation)) :      
                            if userFirstPosition == allPopulation[i][0] :
                                print(userFirstPosition,'인구 수 :',allPopulation[i][1],'명 <순서 상관없이 인구 수가 출력됨>')
                                userFirstPopulation = allPopulation[i][1]
                            elif userSecondPosition == allPopulation[i][0] :
                                print(userSecondPosition,'인구 수 :',allPopulation[i][1],'명 <순서 상관없이 인구 수가 출력됨>')
                                userSecondPopulation = allPopulation[i][1]
                        print("")


                        if userWishRegion == '대구광역시' or userWishRegion == '충청남도' :
                            firstGDPvalue = []
                            firstGDPname = []
                            for i in range(len(allGDP)) :
                                if userFirstPosition == allGDP[i][0] :
                                    firstGDPname.append(allGDP[i][2])   # 컴마(,)로 이루어진 항목 탓에 이름이 잘려나오는 항목이 있음
                                    firstGDPvalue.append(allGDP[i][3])  # 모든 이름을 출력하기 위해 방법 구상 중 // 해결 완료

                            secondGDPvalue = []
                            secondGDPname = []

                            for i in range(len(allGDP)) :
                                if userSecondPosition == allGDP[i][0] :
                                    secondGDPname.append(allGDP[i][2]) 
                                    secondGDPvalue.append(allGDP[i][3])

                        else :     
                            # 도시 별로 GDP 지정
                            firstGDPvalue = []
                            firstGDPname = []

                            for i in range(len(allGDP)) :
                                if userFirstPosition == allGDP[i][0] :
                                    firstGDPname.append(allGDP[i][1])   # 컴마(,)로 이루어진 항목 탓에 이름이 잘려나오는 항목이 있음
                                    firstGDPvalue.append(allGDP[i][2])  # 모든 이름을 출력하기 위해 방법 구상 중 // 해결 완료

                            secondGDPvalue = []
                            secondGDPname = []
                            
                            for i in range(len(allGDP)) :
                                if userSecondPosition == allGDP[i][0] :
                                    secondGDPname.append(allGDP[i][1]) 
                                    secondGDPvalue.append(allGDP[i][2])

                        # 산업 종류 출력
                        print('산업 종류')
                        for i in range(len(firstGDPname)-1) :
                            print(firstGDPname[i],end=' / ')
                        print(firstGDPname[i+1])

                        # 사용자에게 산업을 선택받음
                        while True :
                            print("\n",userFirstPosition, ' ',userSecondPosition,sep ='')
                            print("If you want to back, press 0")
                            userHopeIndustry = input('Press your hope industry :')

                            if userHopeIndustry == '0' : 
                                break

                            else :
                                #두 도시에 지정한 산업의 GDP 값 비교
                                for i in range(len(firstGDPname) - 1) :
                                    comparardGDP = int(firstGDPvalue[i]) - int(secondGDPvalue[i])
                                    if userHopeIndustry == firstGDPname[i] :
                                        if firstGDPvalue[i] == '0' or secondGDPvalue[i] == '0' :
                                            print("There are areas where GDP is zero.")
                                            break                                        
                                        else :
                                            GDPOne = round(int(firstGDPvalue[i])/int(userFirstPopulation),4)
                                            GDPTwo = round(int(secondGDPvalue[i])/int(userSecondPopulation),4)
                                            compareFGDPvalue = 100*round(GDPOne/GDPTwo,4)
                                            compareSGDPvalue = 100*round(GDPTwo/GDPOne,4)
                                            
                                            print(userFirstPosition, userHopeIndustry, 'GDP :', firstGDPvalue[i], '(단위 : 백만원)')
                                            print(userSecondPosition, userHopeIndustry, 'GDP :', secondGDPvalue[i], '(단위 : 백만원)')
                                            print(userFirstPosition, userHopeIndustry, '1인당 GDP :', GDPOne)
                                            print(userSecondPosition, userHopeIndustry, '1인당 GDP :', GDPTwo)

                                            if comparardGDP > 0 :
                                                print("1인당,",userSecondPosition, '대비', userFirstPosition, userHopeIndustry,'GDP :', compareFGDPvalue, '% ', '약', round(compareFGDPvalue/100,2),'배')
                                            else :
                                                print("1인당,",userFirstPosition, '대비', userSecondPosition, userHopeIndustry,'GDP :', compareSGDPvalue, '% ', '약', round(compareSGDPvalue/100,2),'배')
    else : break
```
