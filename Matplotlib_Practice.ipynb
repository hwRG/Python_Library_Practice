```python
# 주피터에서는 그래프를 위해 매직이 필수!
import matplotlib as mpl
import matplotlib.pylab as plt
%matplotlib inline

# mpl로 설치된 폰트를 사용할 수 있게 함
mpl.rc('font', family='12LotteMartDreamMedium')
plt.title("한우")
plt.xlabel("냠냠")
plt.ylabel("념념")
# 각 인덱스를 매핑해 줌, x축이 먼저, y축이 다음
plt.plot([10, 20, 30, 40], [1, 4, 9, 16])
plt.show()
```
